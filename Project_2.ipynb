```python
import os
import pandas as pd

# Function to convert coordinates to bounding boxes
def convert_to_bounding_box(row):
    x0, y0, x1, y1 = row['x0'], row['y0'], row['x1'], row['y1']
    return f'[{x0}, {y0}, {x1}, {y1}]'

# Directory containing CSV files
directory = "/Users/qanand/Documents/DocBank/DocBank_samples/Dataframes"

# Process each CSV file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)
        
        # Read CSV files into a DataFrame
        data = pd.read_csv(file_path)
        
        # Create 'bbox' column from coordinates
        data['bbox'] = data.apply(convert_to_bounding_box, axis=1)
        
        # Drop the original coordinate columns
        data.drop(columns=['x0', 'y0', 'x1', 'y1'], inplace=True)
        
        # Save the modified DataFrame back to the CSV file
        data.to_csv(file_path, index=False)
```
