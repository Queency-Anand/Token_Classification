```python
import pandas as pd
import os
import re

# Directory containing CSV files and images
csv_directory = '/Users/qanand/Documents/DocBank/DocBank_samples/Dataframes'
image_directory = '/Users/qanand/Documents/DocBank/DocBank_samples/DocBank_samples'  # Replace with the actual path to your image directory

# Iterate over each CSV file in the directory
for csv_filename in os.listdir(csv_directory):
    if csv_filename.endswith('.csv'):
        # Read CSV file into a DataFrame
        csv_filepath = os.path.join(csv_directory, csv_filename)
        df1 = pd.read_csv(csv_filepath)
        
        # Extract the numeric part from the CSV filename
        csv_numeric_part = re.search(r'(\d+)', csv_filename).group(1)
        
        # Find the image filename that matches the numeric part of the CSV filename
        image_filename = None
        for image_filename_candidate in os.listdir(image_directory):
            if image_filename_candidate.endswith('ori.jpg'):
                image_numeric_part = re.search(r'(\d+)', image_filename_candidate).group(1)
                if csv_numeric_part == image_numeric_part:
                    image_filename = image_filename_candidate
                    break
        
        # If image filename is found, add it to the "image" column
        if image_filename:
            df1['image'] = image_filename
            # Save DataFrame back to CSV file with the added "image" column
            df1.to_csv(csv_filepath, index=False)
            
```

```python
import pandas as pd
import os

# Directory containing CSV files
csv_directory = '/Users/qanand/Documents/DocBank/DocBank_samples/Dataframes'

# Iterate over each CSV file in the directory
for csv_filename in os.listdir(csv_directory):
    if csv_filename.endswith('.csv'):
        # Read CSV file into a DataFrame
        csv_filepath = os.path.join(csv_directory, csv_filename)
        df = pd.read_csv(csv_filepath)
        
        # Extract the id from the filename before '.csv'
        file_id = csv_filename.split('.')[0]
        
        # Add id column as the first column
        df.insert(0, 'id', file_id)
        
        # Drop unnecessary columns
        df.drop(['Font'], axis=1, inplace=True)
        
        # Rename columns
        df.rename(columns={'Words': 'token', 'bbox': 'bounding_box', 'Label': 'Label'}, inplace=True)
        
    
        
        # Drop redundant columns
        df.drop(['R', 'G', 'B'], axis=1, inplace=True)
        
        # Save the modified DataFrame to the same CSV file, overwriting the original
        df.to_csv(csv_filepath, index=False)

        # Display message
        print(f"Processed DataFrame saved to: {csv_filepath}")
```
